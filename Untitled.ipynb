### Исследование ответа CRM для опредления соотношения количества клиентов и используемых ими веток.

Для получения ответа, использовалась следующая команда. 
Здесь мы выбираем из таблицы `vtiger_potentialscf` только нужные нам поля, а именно регистрационный номер, ветку и версию.

`sqlexec "select cf_regnumber, cf_regbranch, cf_regversion from vtiger_potentialscf where cf_regproduct = 'CRB-Reductor'" > product_list`

Удаляем из него пустые строки и с помощью скрипта ниже парсим данный файл.

```python
#!/usr/bin/env python
"""Generate statistsics for Reductor."""

import csv


def generate_data_for_csv(file):
    """File to list."""
    with open(file) as raw_file:
        lines = [line.replace('\t', ',').replace(
            '\n', '').split(',') for line in raw_file.readlines()]
        return lines


def fill_csv_file():
    """Write data to csv file."""
    with open('statistic_reductor.csv', 'w') as correct_file:
        stat = csv.writer(correct_file)
        stat.writerows(generate_data_for_csv('product_list'))


if __name__ == '__main__':
    fill_csv_file()
```

Открваем csv-файл, при этом именуем столбцы таблицы. Это нам поможет проще к ним обращаться.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
df = pd.read_csv('/home/zimo/statistic_reductor.csv', names=['REGNUM', 'BRANCH', 'VERSION'])
df
```

Выведем полученную таблицу, при этом, отбросив строки с пустыми `BRANCH` и `VERSION` 

```python
df.dropna(thresh=2)
```

Какие есть ветвки

```python
branches = []
for branch in df.dropna(thresh=2)['BRANCH']:
    if branch not in branches:
        branches.append(branch)
branches
```

В количественном соотношении

```python
import matplotlib.pyplot as plt
import numpy as np
data = df.dropna(thresh=2).groupby('BRANCH').size().plot.bar()
```

Сколько клиентов по версиям

```python
df.dropna(thresh=2).groupby('VERSION').size().sort_values(ascending=False)
```
